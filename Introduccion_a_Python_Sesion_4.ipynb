# Matplotlib


Matplotlib es una librería para la creación y manipulación de gráficos 2D y 3D en Python. Está inspirada en las funcionalidades de graficación en MATLAB. 

Para más información consultar https://matplotlib.org/

```python
#importar la librería

import matplotlib.pylab as plt

import numpy as np
```

Hay dos formas de construir gráficos con esta librería en Python:


1. Método funcional
2. Método orientado a objetos

**Método funcional**


En el método funcional, a través de la función `plt.plot`graficamos una o más variables. Veamos un ejemplo de una parábola:

```python
# Gráfiquemos una parabola

x = np.linspace(-10, 10, 30)

print(x)
```

```python
np.shape(x)
```

```python
plt.plot(x)
```

```python
plt.plot(x,x)
```

```python
plt.plot(x, x**2)
```

```python
plt.plot(x**2, x)
```

```python
plt.plot(x, x**2, 'r*')
```

```python
plt.plot(x, x**2)
plt.plot(x, x, 'r*')
plt.xlabel('Este es el eje X')
plt.ylabel('Este es el eje Y')
plt.title('La gráfica de una parábola')
```

```python
plt.subplot(2,1,1)
plt.plot(x, x**2)

plt.subplot(2,1,2)
plt.plot(x, x, 'r*')
```

**2. Método orientado a objetos**




El término 'figure' se refiere per se al espacio de graficación, donde podemos ubicar diferentes 'axes' (lo que entendemos como 'plot'). Los objetos 'axes' pueden de tener 2 o 3 'axis' (3 en el caso de figuras en 3D).

```python
fig, ax = plt.subplots() #Esta línea crea un 'figure' y un 'ax'.

#Ahora utilizamos el ax como objeto y graficamos sobre él

ax.plot(x, x**2)
ax.plot(x, x)

ax.set_xlabel('X axis', fontdict={'fontsize':40, 'color':'g'})
ax.set_ylabel('Y axis')
ax.set_title('Esta es mi parábola')

```

```python
fig, axes = plt.subplots(1,2, figsize=[10,5])

```

```python
fig, axes = plt.subplots(1,3, figsize=[10,5])

axes[0].plot(x, x)
axes[1].plot(x, x**2)
axes[2].plot(x, x**3)

```

**Otros gráficos**



Demos un vistazo de cómo hacer gráficos en 3D con matplotlib:


```python
from mpl_toolkits.mplot3d import Axes3D

fig, ax = plt.subplots()
ax = plt.axes(projection='3d')

ax.plot(x,x**2)
ax.plot(x,x**2,10)
ax.plot(x,x**2,30)

ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
```

```python
x = np.random.rand(1,100)
y = np.random.rand(1,100)
z = np.random.rand(1,100)
```

```python
fig, ax = plt.subplots()
ax = plt.axes(projection='3d')

ax.scatter(x,y,z, c='r')
```

```python
fig, ax = plt.subplots(figsize=[20,20])
ax = plt.axes(projection='3d')

x = np.arange(-100,100)
y = np.arange(-100,100)

X, Y = np.meshgrid(x,y)

Z = X**2 + Y**2

ax.plot_surface(X, Y, Z, cmap='seismic')


```

```python
fig, ax = plt.subplots(figsize=[20,20])
ax = plt.axes(projection='3d')

x = np.arange(-100,100)
y = np.arange(-100,100)

X, Y = np.meshgrid(x,y)

Z = X**2 + Y**2

ax.plot_wireframe(X, Y, Z, cmap='seismic')
```

# Entrada y salida de archivos (input - output)

Hasta ahora, hemos visto únicamente ejemplos que usan entradas del usuario con el teclado y como salida se imprime en pantalla el resultado. Aunque esto es útil para resolver tareas pequeñas, en la vida real se requiere utilizar grandes cantidades de datos, p. ej. LAS, Zmap, archivos que provienen de equipos, entre otros. Veamos algunos ejemplos de como abrir, leer y escribir sobre archivos de texto plano.

**Nota:** En muchos varios módulos de Python existen codigos para abrir distintos tipos de datos p. ej. formato MATLAB, excel, segy, etc... Por ahora solo trabajaremos con archivos de texto plano.

```python
fname = "Test.txt"

f = open(fname, "w")#Abrimos el archivo
```

```python
f.write('Primera linea \n')
f.write('Segunda linea \n')
f.write('Tercera linea \n')
f.write('Cuarta linea \n')


f.close() #Cerramos el archivo
```

```python
print('----Leyendo linea a linea \n')

fname = "Test.txt"

f = open(fname,'r')

for line in f:
    print(line, end='')

f.close()
```

```python
print('----Leyendo como lista \n')

f = open(fname, 'r')
f_list = f.readlines()

print(f_list)
print(f_list[1])
f.close()
```

```python
print('----Leer todo el archivo \n')

f = open(fname, 'r')
text = f.read()

print(text)
```
